```python
# rerun the loop with a correct save method
from pathlib import Path
import subprocess, sys, shutil

chunk_dir = Path("/content/netsolp_chunks")
drive_out_dir = Path("/content/drive/MyDrive/SynBio Software/data/processed/netsolp_chunks")
drive_out_dir.mkdir(parents=True, exist_ok=True)

pred_dir = Path("/content/netsolp/Data_and_Code/PredictionServer")
predict_py = pred_dir / "predict.py"

NUM_THREADS = "2"

chunks = sorted(chunk_dir.glob("chunk_*.fasta"))
print("Found chunks:", len(chunks))

for fasta in chunks:
    idx = fasta.stem.split("_")[1]          # "000"
    local_csv = Path(f"/content/netsolp_chunk_{idx}.csv")
    drive_csv = drive_out_dir / f"netsolp_chunk_{idx}.csv"

    # skip if already on Drive
    if drive_csv.exists() and drive_csv.stat().st_size > 0:
        print(f"âœ… Skip chunk {idx} (already exists on Drive)")
        continue

    cmd = [
        sys.executable, str(predict_py),
        "--FASTA_PATH", str(fasta),
        "--OUTPUT_PATH", str(local_csv),
        "--MODEL_TYPE", "Distilled",
        "--PREDICTION_TYPE", "U",
        "--NUM_THREADS", NUM_THREADS,
    ]

    print(f"\nâ–¶ Running chunk {idx}: {fasta.name}")
    r = subprocess.run(cmd, cwd=str(pred_dir))
    if r.returncode != 0:
        raise RuntimeError(f"Chunk {idx} failed with return code {r.returncode}")

    # copy to Drive (works across /content -> /content/drive)
    shutil.copy2(local_csv, drive_csv)
    print(f"ðŸ’¾ Saved: {drive_csv}  (bytes={drive_csv.stat().st_size})")

    # optional: free local disk
    try:
        local_csv.unlink()
    except FileNotFoundError:
        pass
```
