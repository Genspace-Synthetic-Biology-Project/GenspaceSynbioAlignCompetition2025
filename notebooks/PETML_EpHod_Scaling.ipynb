```python
# Define the pH penalty (bell curve)
import numpy as np

def ph_scale(ph_opt, target_ph, sigma=1.0):
    """
    Returns a number between 0 and 1.
    1.0 if target_ph == ph_opt.
    Smaller as you move away from the optimum.
    sigma controls how quickly activity falls off.
    """
    return np.exp(-0.5 * ((target_ph - ph_opt) / sigma) ** 2)
```

```python
# Create activity_1 and activity_2
sigma = 1.0  # start here; later you can try 1.5 or 2.0

df["activity_1_raw"] = df["petml_scores"] * ph_scale(df["Ensemble"], 5.5, sigma=sigma)
df["activity_2_raw"] = df["petml_scores"] * ph_scale(df["Ensemble"], 9.0, sigma=sigma)

```

```python
# Are the scalers sane?
df["scale_5p5"] = ph_scale(df["Ensemble"], 5.5, sigma=sigma)
df["scale_9p0"] = ph_scale(df["Ensemble"], 9.0, sigma=sigma)

df[["Ensemble","scale_5p5","scale_9p0"]].describe()
```
