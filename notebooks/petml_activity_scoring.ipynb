````python
# write every row’s protein sequence to a FASTA file named `predictive_pet_zero_shot_2025.fasta`
# FASTA headers become `>seq_0`, `>seq_1`, etc

def df_to_fasta(df, fasta_path, seq_col="sequence", id_prefix="seq"):
    with open(fasta_path, "w") as f:
        for i, row in df.iterrows():
            f.write(f">{id_prefix}_{i}\n")
            f.write(f"{row[seq_col]}\n")

# # Wrote FASTA to: /content/drive/MyDrive/SynBio Software/data/processed/predictive_pet_zero_shot_2025.fasta
````

<!-- You should write a mapping file **now** because:

> **PETML will forget everything about your original table except the FASTA header.**

Once PETML finishes, the *only* thing linking its scores back to your data is the FASTA ID (`seq_1234`).
If you don’t save what `seq_1234` means **before** running PETML, you are trusting your memory, notebook state, and index order not to break later.

The mapping file is a **receipt**:

> “This FASTA ID came from this row.”



## What PETML actually sees (and what it does not)

Right now, PETML will be given **only this**:

```text
>seq_1732
AADNPYQRGPDPTNASIEAATGPF...
```

PETML does **not** know:

* where this sequence came from
* what row it was in your CSV
* what its activity columns were
* what expression value it had
* what mutation it represents

After PETML runs, it will output something like:

```csv
seq_id,petml_scores
seq_1732,2.481
```

That’s all you get.

So the question becomes:

> When you see `seq_1732` in a CSV **days or weeks from now**, how do you know which original sequence that was?



## “But I *know* seq_1732 is row 1732…”

You know that **right now**.

But here’s what commonly breaks that assumption:

### 1️⃣ Reordering happens silently

Any of these will break the index ↔ sequence relationship:

```python
df = df.sample(frac=1)        # shuffle
df = df.sort_values("score") # reorder
df = df.dropna()              # remove rows
df.reset_index(drop=True)    # renumber rows
```

After that:

* `seq_1732` ≠ row 1732 anymore
* and you will not get an error — just wrong merges



### 2️⃣ Colab sessions reset

Colab will eventually:

* restart the kernel
* disconnect
* lose in-memory variables

If that happens and all you have is:

* a FASTA file
* a PETML `final_scores.csv`

you have **no guaranteed way** to reconstruct the original row order.



### 3️⃣ Future-you is a different person

This is not a joke — it’s a real failure mode.

Six weeks from now:

* you open `final_scores.csv`
* you see `seq_2871`
* you ask: *“what exactly was this variant again?”*

The mapping file answers that instantly.



## What the mapping file actually gives you

The mapping file is tiny, cheap, and boring — but powerful:

```csv
seq_id
seq_0
seq_1
seq_2
...
```

Or better:

```csv
seq_id,sequence
seq_0,AADNPYQRGPDPTNASIEAATGPF...
seq_1,AADNPYQRGPDPTNASIEAATGPFC...
```

This means:

* PETML outputs → `seq_1732`
* mapping file → tells you the exact sequence
* sequence → tells you the mutation
* mutation → tells you which experimental row it came from

All chains are intact.



## Why **before**, not after

Once PETML has run:

* you cannot recover this mapping if anything went wrong
* you cannot reassign IDs safely
* you cannot prove correctness

Before PETML:

* mapping is trivial
* guaranteed correct
* zero compute cost

After PETML:

* mapping reconstruction ranges from annoying → impossible

This is why the advice is **do it now**.



## Analogy (non-bio, very concrete)

Imagine shipping 4,988 unlabeled boxes to a warehouse.

You *think*:

> “Box #1732 is the one with the blue shoes.”

But you don’t write it down.

The warehouse sends back:

> “Box #1732 scored highest quality.”

Now you want to know which shoes those were.

The mapping file is the shipping manifest.



## Bottom line

You are about to spend:

* **time** (PETML runtime)
* **compute**
* **attention**

The mapping file costs:

* **seconds**
* **no compute**
* **zero downside**

It turns a fragile workflow into a robust one. -->
